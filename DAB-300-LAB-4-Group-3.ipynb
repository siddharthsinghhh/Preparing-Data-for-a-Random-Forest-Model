```python
# The below code is from class notebooks part-1 missing_data
def sniff_modified(df):
    with pd.option_context("display.max_colwidth", 20):
        info = pd.DataFrame()
        info['data type'] = df.dtypes
        info['percent missing'] = df.isnull().sum()*100/len(df)
        info['No. unique'] = df.apply(lambda x: len(x.unique()))
        info['unique values'] = df.apply(lambda x: x.unique())
        return info.sort_values('data type')
```

```python
# Imputing null values with mode
def impute_nan_most_frequent_category(dataFrame,ColumnName):

     most_frequent_category=dataFrame[ColumnName].mode()[0]

     dataFrame[ColumnName + "_Imputed"] = dataFrame[ColumnName]
     dataFrame[ColumnName + "_Imputed"].fillna(most_frequent_category,inplace=True)

for Columns in ['Unique Label?','Label Recalc?']:
    impute_nan_most_frequent_category(data,Columns)
```

### Transforming the data
- Now we have dropped around 20 columns with 100% null values which are of no use for the model.
- As per the instructions we have dropped columns related to fuel economy except our target variable which is 'comb unadj FE'.
- Next we have dropped description columns as they are of no use to the target. Moreover they are just description of other columns. For example there are columns called 'Air Aspir Method' which have values of different aspiration methods in short form and 'Air Aspir Method Desc' has the long form of the methods mentioned in the earlier columns.
- Next we have dropped Passenger and luggage volume columns. Though they are important to target variable, these columns have more than 60% null values and there is no logical way to impute them. Because there are columns where all 2WD pass & lugg, 4WD pass & lugg, and Hatchback pass & lugg are null values. if that was not the case we could have assumed that 2WD pass & lugg values are nulls because they are either 4WD or hatchback vehicles.
- Next we dropped columns related to batteries, regen, and fuel2 because there are around 90% null values in these columns and there is no possible way to impute the nulls. Moreover fuel2 columns which are columns related to alternate fuel are dropped because we are trying find out the fuel economy of conventional fuel not alternate fuel.
- Guzzler is a component in automobile which causes heavy fuel consumption and related to fuel economy. But there are more than 90% null values and cannot logically impute them.
- Range1 - Model Type Driving Range - Conventional Fuel, Relabel, # Drive Motor Gen','Motor Gen Type, If Other','Rated Motor Gen Power (kW)','Fuel Cell Vehicle (Y or N)','Off Board Charge Capable (Y or N)','GHG 1-10 rating on Ethanol (EPA Determined)','#2 Smog Rating Test Group','#2 Mfr Smog Rating (Mfr Smog 1-10 Rating on Label for Test Group 2)','Unnamed: 0'
- These columns are either unrelated like fuel unit or have more null values which cannot be replaced logically.
- Transmission column is removed because there are columns called Trans and Gears which says the same information. For example there is a value in Transmission which says Auto(S8) which means semi automatic vehicle with 8 gears. If we compare with Trans and Gears, the transmission values is divided as SA in Trans and 8 in Gears.
- Mfr Name is dropped because there is other column called verify mfr code which is a short of the Mfr name.

- Model Year, Suppressed, and Police/Emerg are dropped because they have only one unique value.

- Earlier we have dropped description columns because every description column has the parent column, but Calc Approach Desc has not parent column so we again added that column. Otherwise it will cause loss of information.

- In Air Aspir Method there is a value called Naturally Aspirated(from its description column) and this value is written in short form 'NA'. And this NA values are considered as null values in python. So we replaced the nulls with NAT.

- Max Ethanol and Max Biodiesel are linked with type of fuel used in the vehicles.  Gasoline has the chemical component ethanol and diesel has the chemical component biodiesel. We have checked that where ever there are null values in ethanol the fuel used in diesel and biodiesel has the value. So we replaced nulls with 0 in both ethanol and biodiesel.

- There are 2 columns which says how much the person saved on fuel in last 5 years and how much the person spent extra on fuel in last 5 years. So here we have seen that where ever there are nulls in "$ You Save over 5 years (amount saved in fuel costs over 5 years - on label)', there are values in '$ You Spend over 5 years (increased amount spent in fuel costs over 5 years - on label) '. So we replaced the nulls in first column with 0 and created a new column called whether the amount is saved or not. If there are 0 in $ You Save over 5 years (amount saved in fuel costs over 5 years - on label) then not saved else saved. And we dropped the original columns.

- Next we seperated categorical and numerical variables. In category variables the column 'Car/Truck Category - Cash for Clunkers Bill.' Have less null values we cannot replace them so we dropped it.

- Using categorical imputation method we have imputed nulls in 'Unique Label?' and 'Label Recalc?' with mode of their respective columns.

- Using Label encoding we have encoded categorical variables. This method converts values in categorical in numbers from 0.

- In numeric variables, first we checked for outliers. Then using imputer function from sklearn we have replaced nulls with their mean. Because there are not much outliers in these columns.
- Finally we have got the dataset of 1290*41 after the data preprocessing. And everything is in numerical data type.
